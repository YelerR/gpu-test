```python
import numpy as np
import tensorflow as tf
from tensorflow.keras import datasets, layers, models

GPU_ON = tf.debugging.set_log_device_placement(True)
print(GPU_ON)
```

```python
from tensorflow.python.client import device_lib
device_lib.list_local_devices()
```

```python
# 데이터 불러오기
(x_train, y_train), (x_test, y_test) = datasets.mnist.load_data()
```

```python
# 넘파이 데이터를 텐서 데이터로 변환
x_train = tf.convert_to_tensor(x_train, dtype=tf.float32)
x_test = tf.convert_to_tensor(x_test, dtype=tf.float32)
y_train = tf.one_hot(y_train, depth=len(np.unique(y_train)))
y_test = tf.one_hot(y_test, depth=len(np.unique(y_train)))
```

```python
# 레이어 설계
model = tf.keras.Sequential()
model.add(tf.keras.layers.Flatten())
model.add(tf.keras.layers.Dense(100, activation='relu'))
model.add(tf.keras.layers.Dense(100, activation='relu'))
model.add(tf.keras.layers.Dense(10, activation='softmax'))
```

```python
# 모델 컴파일
model.compile(loss=tf.keras.losses.categorical_crossentropy, optimizer=tf.keras.optimizers.SGD(), metrics=['accuracy'])
```

```python
from tensorflow.python.client import device_lib
device_lib.list_local_devices()
```

```python
# CPU 학습
print("CPU를 사용한 학습")
with tf.device("/device:CPU:0"):
    model.fit(x_train, y_train, batch_size=512, epochs=3)
```

```python
print("GPU를 사용한 학습")
with tf.device("/device:GPU:0"):
    model.fit(x_train, y_train, batch_size=512, epochs=3)
```

```python
print("GPU를 사용한 학습")
with tf.device("/device:GPU:1"):
    model.fit(x_train, y_train, batch_size=512, epochs=3)
```
